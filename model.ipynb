```python
import pandas as pd
from snowflake_info import df

print(df.head())
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize = (2, 1))
scatter = plt.scatter(
                    x=df['SOCK_AGE_DAYS'],
                    y=df['TARGET_VANISHED'],
                    c=df['COLOR_INTENSITY'].astype(float),
                    s=df['WASHER_SPIN_RPM'].astype(float),
                    cmap='viridis',
                    alpha=0.7)

plt.colorbar(scatter, label = 'input 2 mag')

plt.xlabel('input 1')
plt.ylabel('output')
plt.title('Scatter Plot using X, Y, Color, and Size')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

```
