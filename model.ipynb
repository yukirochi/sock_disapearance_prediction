```python
import pandas as pd
import numpy as np
from snowflake_info import df

df = df.sort_values(by='SOCK_AGE_DAYS').head(100)

# print(df.head())

df_train = df.iloc[:50, :]
df_test = df.iloc[50:, :]

sock_days_train = np.array(df_test['SOCK_AGE_DAYS'].astype(float))
wash_temp_train = np.array(df_test['WASH_TEMP_C'].astype(float))
dryer_train = np.array(df_test['DRYER_LINT_GRAMS'].astype(float))

sock_days_test = np.array(df_test['SOCK_AGE_DAYS'].astype(float))
wash_temp_test = np.array(df_test['WASH_TEMP_C'].astype(float))
dryer_test = np.array(df_test['DRYER_LINT_GRAMS'].astype(float))

x_train = wash_temp_train + sock_days_train + dryer_train
y_train = np.array(df_train['TARGET_VANISHED'])

x_test = wash_temp_test + sock_days_test + dryer_test
y_test = np.array(df_train['TARGET_VANISHED'])

m, b = np.polyfit(x_train, y_train, 1)

model =  lambda x : (m * x) + b

predicted = model(x_test)

residual = y_test - predicted

print(residual)

corrects_prediction = np.abs(residual ) <= 0.7


accuracy = np.mean(corrects_prediction)

print(accuracy)
```
